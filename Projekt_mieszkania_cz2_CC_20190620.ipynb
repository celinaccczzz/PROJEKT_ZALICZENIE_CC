## 2. Sprawdzenie kolumny "liczba_pieter"

```python
df_valuable = pd.read_csv("mieszkania_etap2_df.csv")
# oraz od razu usuwam: Unnamed
df_valuable = df_valuable.drop(["Unnamed: 0"], axis=1)
```

```python
#sprawdzam jakie sa unikalne wartosci w kol o indexie "liczba_pieter" :
print(f'Unique Values for \"liczba_pieter\" to: {df_valuable["liczba_pieter"].unique()}')
print(f'liczba wartosci unikalnych to: {len(df_valuable["liczba_pieter"].unique())}')
print(f'liczba wartosci wypelnionych - czyli nie NAN w kolumnie to: \n {df_valuable.liczba_pieter.value_counts()}')
print(f'liczba missingow w kolumnie to: {df_valuable["liczba_pieter"].isnull().sum()}')

# WNIOSEK: trzeba pogrupowac, patrzac na wykres - oceniam, ze dobre beda kategorie: 
#do 5 pietra, do 10 pietra i pow. 10 pietra
# sa NANy - ktore wypelnie wartoscia najczestsza
```

```python
# zamieniam NANy na wartosc najczestsza - czyli na 4 pietro
# grupa NANow wpadnie do 4 pietra - co pasuje do zamierzonego grupowania na 3 koszyki 
#(koszyk do 5 pietra jest najliczniejszy)
imp = SimpleImputer(missing_values= np.nan, strategy='most_frequent')
df_valuable['liczba_pieter']= imp.fit_transform(df_valuable[['liczba_pieter']]).ravel()
df_valuable['liczba_pieter'].value_counts()
```

```python
#sprawdzam PONOWNIE jakie sa unikalne wartosci w kol o indexie "liczba_pieter" :
print(f'Unique Values for \"liczba_pieter\" to: {df_valuable["liczba_pieter"].unique()}')
print(f'liczba wartosci unikalnych to: {len(df_valuable["liczba_pieter"].unique())}')
print(f'liczba wartosci wypelnionych - czyli nie NAN w kolumnie to: \n {df_valuable.liczba_pieter.value_counts()}')
print(f'liczba missingow w kolumnie to: {df_valuable["liczba_pieter"].isnull().sum()}')
```

```python
df_valuable['liczba_pieter']=df_valuable['liczba_pieter'].astype(float)
```

```python
# WNIOSEK: patrzac na wykres - oceniam, ze dobre beda kategorie: 
# do 5 pietra, do 10 pietra i pow. 10 pietra
# grupuje na 3 koszyki: 1, 2, 3
df_valuable.loc[df_valuable['liczba_pieter'] < 6, 'liczba_pieter'] = '1'
df_valuable.loc[df_valuable['liczba_pieter'] == 6, 'liczba_pieter'] = '2'
df_valuable.loc[df_valuable['liczba_pieter'] == 7, 'liczba_pieter'] = '2'
df_valuable.loc[df_valuable['liczba_pieter'] == 8, 'liczba_pieter'] = '2'
df_valuable.loc[df_valuable['liczba_pieter'] == 9, 'liczba_pieter'] = '2'
df_valuable.loc[df_valuable['liczba_pieter'] == 10, 'liczba_pieter'] = '2'
df_valuable['liczba_pieter']=df_valuable['liczba_pieter'].astype(float)
df_valuable.loc[df_valuable['liczba_pieter'] > 10, 'liczba_pieter'] = '3'
```

```python
df_valuable['liczba_pieter']=df_valuable['liczba_pieter'].astype(int)
```

```python
print(f'Unique Values for \"liczba_pieter\" to: {df_valuable["liczba_pieter"].unique()}')
print(f'liczba wartosci unikalnych to: {len(df_valuable["liczba_pieter"].unique())}')
print(f'liczba wartosci wypelnionych - czyli nie NAN w kolumnie to: \n {df_valuable.liczba_pieter.value_counts()}')
print(f'liczba missingow w kolumnie to: {df_valuable["liczba_pieter"].isnull().sum()}')
```
